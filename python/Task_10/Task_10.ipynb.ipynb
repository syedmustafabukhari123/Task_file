### Imports and load environment variables

```python
import asyncio
import os
import random
from dotenv import load_dotenv

load_dotenv() 
```

### Define Mock Data Source Class

```python
class MockSource:
    def __init__(self, name, api_key_env_var):
        self.name = name
        self.api_key = os.getenv(api_key_env_var)

    async def fetch_data(self):
        await asyncio.sleep(random.uniform(0.5, 2.0))
        return {self.name: [random.randint(1, 100) for _ in range(5)]}
```

### Merge and Summarize Functions

```python
def merge_data(data_list):
    merged = {}
    for data in data_list:
        for key, values in data.items():
            if key not in merged:
                merged[key] = []
            merged[key].extend(values)
    return merged

def summarize_data(merged_data):
    summary = {}
    for key, values in merged_data.items():
        summary[key] = {
            "count": len(values),
            "min": min(values),
            "max": max(values),
            "average": round(sum(values) / len(values), 2)
        }
    return summary
```

### Main Async Runner

```python
async def main():
    sources = [
        MockSource("Source1", "SOURCE1_API_KEY"),
        MockSource("Source2", "SOURCE2_API_KEY"),
        MockSource("Source3", "SOURCE3_API_KEY"),
    ]
    tasks = [source.fetch_data() for source in sources]
    results = await asyncio.gather(*tasks)
    merged = merge_data(results)
    summary = summarize_data(merged)
    for source, stats in summary.items():
        print(f"{source}: {stats}")
```

### Run the Async Dashboard


```python
await main()
```
