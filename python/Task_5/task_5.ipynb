### Order Discount Pipeline

---

### Raw Data of Customer 

```python
from functools import reduce
orders = [
    {"order_id": 1, "customer": "Ali", "category": "electronics", "price": 1200},
    {"order_id": 2, "customer": "Sara", "category": "clothing", "price": 150},
    {"order_id": 3, "customer": "John", "category": "electronics", "price": 800},
    {"order_id": 4, "customer": "Emma", "category": "grocery", "price": 60},
    {"order_id": 5, "customer": "David", "category": "clothing", "price": 400},
]
```

### Manages Discounts and Filter Orders Below Minimum Threshold

```python
discount_rules = {
    "electronics": 0.10,   
    "clothing": 0.20,     
    "grocery": 0.05        
}

discounted_orders = list(
    map(
        lambda order: {
            **order,
            "final_price": order["price"] * (1 - discount_rules.get(order["category"], 0))
        },
        orders
    )
)

discounted_orders
```

```python
minimum_threshold = 100

valid_orders = list(
    filter(
        lambda order: order["final_price"] >= minimum_threshold,
        discounted_orders
    )
)

valid_orders
```

### Sort Orders by Revenue

```python
sorted_orders = sorted(
    valid_orders,
    key=lambda order: -order["final_price"]
)

sorted_orders
```

### Total Revenue

```python
total_revenue = reduce(
    lambda total, order: total + order["final_price"],
    sorted_orders,
    0
)

total_revenue
```

### Summary

```python
summary = {
    "processed_orders": sorted_orders,
    "total_orders": len(sorted_orders),
    "total_revenue": total_revenue
}

summary
```
