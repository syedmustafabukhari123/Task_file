### Imports and Setup

```python
import os
import time
from datetime import datetime, timedelta
import math
from scanner import list_files, get_file_info, flag_old_files
```

### Setup Folder and Timestamp

```python
folder_path = "TestProject"  
now = datetime.now()
timestamp = now.strftime("%Y%m%d_%H%M%S")  

print(f"Scanning folder: {folder_path}")
print(f"Reports will have timestamp: {timestamp}")
```

### List Files and Flag Old Files

```python
all_files = list_files(folder_path)
old_files = flag_old_files(folder_path, 30)

print(f"Total files in folder: {len(all_files)}")
print("Files older than 30 days:", old_files)
```

### Display Audit Report in Notebook

```python
print("AUDIT REPORT")
for file_name in all_files:
    file_path = os.path.join(folder_path, file_name)
    info = get_file_info(file_path)
    
    print(f"Name          : {info['name']}")
    print(f"Size (KB)     : {info['size_kb']}")
    print(f"Type          : {info['type']}")
    print(f"Last Modified : {info['last_modified'].strftime('%Y-%m-%d %H:%M:%S')}")
    
    if file_name in old_files:
        print("⚠ FLAG       : Not modified in last 30 days!")
    
    print("-"*40)
```

### Generate Penalties Report

```python
penalties_report_path = os.path.join(folder_path, f"penalties_report_{timestamp}.txt")

with open(penalties_report_path, "w") as f:
    f.write("PENALTIES REPORT\n")
    f.write(f"Generated on: {now.strftime('%Y-%m-%d %H:%M:%S')}\n\n")
    
    for file_name in old_files:
        file_path = os.path.join(folder_path, file_name)
        last_mod = get_file_info(file_path)["last_modified"]
        days_overdue = (now - last_mod).days - 30
        fine = math.ceil(days_overdue * 1.0) 
        
        f.write(f"File: {file_name}\n")
        f.write(f"Days overdue: {days_overdue}\n")
        f.write(f"Fine: ${fine}\n")
        f.write("-"*40 + "\n")

print(f"Penalties report saved: {penalties_report_path}")
```
