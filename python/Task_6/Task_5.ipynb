### Server Log Error Streamer
---
### Problem Statement
Modern backend servers generate extremely large log files. These files can become so large that loading them entirely into memory may cause memory issues or slow performance
### Log File

```python
log_data = [
"INFO Server started",
"ERROR Database connection failed",
"WARNING High memory usage",
"CRITICAL Disk failure",
"ERROR Timeout occurred",
"CRITICAL Network failure",
"INFO User logged in",
"ERROR Database connection failed"
]
with open("server_log.txt","w") as file:
    for line in log_data:
        file.write(line + "\n")
```

### Generator to Stream Log File

```python
def read_log_stream(filename):
    
    with open(filename,"r") as file:
        
        for line in file:
            yield line.strip()
```

### Filter Only Critical Errors

```python
def critical_errors(log_stream):
    
    for line in log_stream:
        
        if "ERROR" in line or "CRITICAL" in line:
            yield line
```

### Live Running Count

```python
def count_errors(error_stream):
    
    counts = {}
    
    for error in error_stream:
        
        error_type = error.split()[0]
        
        if error_type not in counts:
            counts[error_type] = 0
        
        counts[error_type] += 1
        
        print("New Error:", error)
        print("Current Count:", counts)
        print()
```

### Run the Streaming System

```python
log_stream = read_log_stream("server_log.txt")

error_stream = critical_errors(log_stream)

count_errors(error_stream)
```
