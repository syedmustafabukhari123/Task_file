### API Middleware Simulator using Decorators and Logging
---
### Import Modules

```python
import logging
import time
from functools import wraps
```

### Configure Logging

```python
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filename="api_activity.log"
)

logger = logging.getLogger("api_middleware")
```

### Execution Timing Decorator

```python
def timing_decorator(func):

    @wraps(func)
    def wrapper(*args, **kwargs):

        start = time.time()

        result = func(*args, **kwargs)

        end = time.time()

        execution_time = end - start

        logger.info(f"{func.__name__} executed in {execution_time:.4f} seconds")

        return result

    return wrapper
```

### Input,Output Logging Decorator

```python
def log_io(func):

    @wraps(func)
    def wrapper(*args, **kwargs):

        logger.info(f"Function {func.__name__} called with args={args}, kwargs={kwargs}")

        result = func(*args, **kwargs)

        logger.info(f"Function {func.__name__} returned {result}")

        return result

    return wrapper
```

### Role-Based Access Control Decorator

```python
def require_role(role):

    def decorator(func):

        @wraps(func)
        def wrapper(user_role, *args, **kwargs):

            if user_role != role:
                logger.warning(f"Access denied for role: {user_role}")
                return "Access Denied"

            logger.info(f"Access granted for role: {user_role}")

            return func(user_role, *args, **kwargs)

        return wrapper

    return decorator
```

```python
@timing_decorator
@log_io
@require_role("admin")
def get_user_data(user_role, user_id):
    time.sleep(1)
    return {"user_id": user_id, "name": "Mustafa"}
```
