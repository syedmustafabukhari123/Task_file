### Library Management System
---
 This notebook implements a simple Library Management System using object oriented programming concepts. The system manages books, members, and borrowing records. Members can borrow and return books, and fines are applied for overdue returns.
### Importing Required Modules

```python
from abc import ABC, abstractmethod
from datetime import datetime
```

### Book Class
 The Book class represents a book in the library.

```python
class Book:
    def __init__(self,title):
        self.title=title
        self.available=True

    def __str__(self):
        return self.title

    def __eq__(self,other):
        return self.title==other.title

    def __hash__(self):
        return hash(self.title)

    def __add__(self,other):
        return self.title+" & "+other.title
```

### Borrowable Abstract Class
This abstract class defines the required methods that every member type must implement for borrowing and returning books.

```python
class Borrowable(ABC):
    @abstractmethod
    def borrow(self,book):
        pass
    @abstractmethod
    def return_book(self,book):
        pass
```

### Member base class 

```python
class Member(Borrowable):
    def __init__(self,name,limit):
        self.name=name
        self.limit=limit
        self.borrowed={}
    def borrow(self,book):
        if len(self.borrowed)<self.limit and book.available:
            book.available=False
            self.borrowed[book]=datetime.now()
    def return_book(self,book):
        if book in self.borrowed:
            book.available=True
            self.borrowed.pop(book)
    def __len__(self):
        return len(self.borrowed)
```

### Membership Tiers using Inheritance

Different membership levels are created using inheritance.

- BasicMember demonstrates single inheritance

- PremiumMember demonstrates multilevel inheritance

- EliteMember demonstrates multiple inheritance

```python
class BasicMember(Member):
    def __init__(self,name):
        super().__init__(name,2)
class PremiumMember(BasicMember):
    def __init__(self,name):
        super().__init__(name)
        self.limit=5
class SpecialAccess:
    def reserve(self,book):
        return book.title+" reserved"
class EliteMember(PremiumMember,SpecialAccess):
    def __init__(self,name):
        super().__init__(name)
        self.limit=10
```

### Library Class
- The Library class manages books and members.

```python
class Library:
    def __init__(self):
        self.books=[]
        self.members=[]

    def add_book(self,book):
        self.books.append(book)

    def add_member(self,member):
        self.members.append(member)

    def borrow(self,*args):
        member=args[0]
        book=args[1]
        member.borrow(book)
```

### Testing the System
- This section creates books and members and demonstrates borrowing functionality.

```python
b1=Book("Python")
b2=Book("Algorithms")

m1=BasicMember("Ali")
m2=EliteMember("Sara")

lib=Library()
lib.add_book(b1)
lib.add_book(b2)

lib.add_member(m1)
lib.add_member(m2)

lib.borrow(m1,b1)

print(len(m1))
print(b1+b2)
```
