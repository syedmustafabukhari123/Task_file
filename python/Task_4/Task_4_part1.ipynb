### E-Commerce Inventory Analyzer Task
--- 
### Problem Statement
 Clean product data, group by category, find the top 5 most expensive items per category, and produce a summary dictionary, using comprehensions extensively.
### Data Setup


```python
messy_products = [
    ('Laptop Pro', 'Electronics', 1200),
    ('Mouse Pad XL', 'Accessories', 15),
    ('T-Shirt Cotton', 'Apparel', 25),
    ('Laptop Pro', 'Electronics', 1200), 
    ('Wireless Mouse', 'Electronics', 50),
    ('Jeans Slim', 'Apparel', 60),
    ('Keyboard Mech', 'Electronics', 150),
    ('Mug Heat-Sensitive', 'Home Goods', 20),
    ('Socks Pack', 'Apparel', None), 
    ('Smartwatch X', 'Electronics', 300),
    ('Desk Lamp LED', 'Home Goods', 45),
    ('Headphones ANC', 'Electronics', 250),
    ('T-Shirt Cotton', 'Apparel', 30), 
    ('Webcam HD', 'Electronics', 90),
    ('Blender Power', 'Home Goods', 110),
    ('Sweater Wool', 'Apparel', 75),
    ('External SSD 1TB', 'Electronics', 180),
    ('Novel Best Seller', 'Books', 15),
    ('Coffee Maker Auto', 'Home Goods', 150),
    ('Monitor 4K', 'Electronics', 450),
    ('Jeans Skinny', 'Apparel', 65),
    ('Book Sci-Fi', 'Books', 12),
    ('Ebook Reader', 'Electronics', 100),
]
```

###  Cleaning and Deduplication

```python
cleaned_products_map = {}
for name, category, price in messy_products:
    if price is not None:
        if name not in cleaned_products_map or price > cleaned_products_map[name]['price']:
            cleaned_products_map[name] = {'category': category, 'price': price}
processed_list = [{'name': name, **details} for name, details in cleaned_products_map.items()]

print(f"--- Processed List Size (Deduplicated & Validated): {len(processed_list)} ---")
```

### Grouping by Category

```python
grouped_by_category = {
    category: [
        prod for prod in processed_list if prod['category'] == category
    ]
    for category in {prod['category'] for prod in processed_list} 
}

print(f"--- Categories Found: {list(grouped_by_category.keys())} ---")
```

### Finding Top 5 Most Expensive per Category

```python
top_5_per_category = {
    category: sorted(
        products,
        key=lambda item: item['price'], 
        reverse=True                   
    )[:5]                              
    for category, products in grouped_by_category.items()
}

print("\n--- Top 5 Most Expensive Items Per Category ---")
for cat, items in top_5_per_category.items():
    print(f"\nCategory: {cat}")
    for item in items:
        print(f"  - {item['name']}: ${item['price']}")
```

### Final Summary

```python
inventory_summary = {
    category: {
        "count": len(products),
        "total_value": sum(item['price'] for item in products), 
        "top_5_expensive": top_5_expensive 
    }
    for category, products in grouped_by_category.items()
    for top_5_expensive in [top_5_per_category.get(category, [])]
}
print("\n--- Final Inventory Summary Dictionary ---")
import json
print(json.dumps(inventory_summary, indent=4))
```
