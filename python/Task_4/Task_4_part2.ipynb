### HR Employee Ranker Task
 
---

### Problem statement 
 Rank employees based on Performance Score (descending), then Experience (descending), then Name (ascending), and implement a fast search function (Binary Search) on the final ranked list.

### Data Setup

```python
employee_data = [
    {'name': 'Alice', 'score': 95, 'experience': 5},
    {'name': 'Bob', 'score': 88, 'experience': 7},
    {'name': 'Charlie', 'score': 95, 'experience': 5},
    {'name': 'David', 'score': 95, 'experience': 8}, 
    {'name': 'Eve', 'score': 88, 'experience': 7},   
    {'name': 'Frank', 'score': 75, 'experience': 3},
    {'name': 'Grace', 'score': 99, 'experience': 10},
]
```

### Sorting according to the Given Conditions

```python
ranked_employees = sorted(
    employee_data, key=lambda emp: (-emp['score'], -emp['experience'], emp['name'])
)

print("--- Ranked Employee List (Top to Bottom) ---")

for index, emp in enumerate(ranked_employees, 1):
    print(f"{index:2}. Score: {emp['score']:3}, Exp: {emp['experience']:2}, Name: {emp['name']}")
```

```python
def binary_search_by_name(arr, target_name):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        mid_name = arr[mid]['name']

        if mid_name == target_name:
            return arr[mid]  
        elif mid_name < target_name:
            low = mid + 1 
        else:
            high = mid - 1 

    return None 
search_name_1 = "David"
search_name_2 = "Eve"
search_name_3 = "Zoe (Non-existent)"

print(f"\n--- Searching for '{search_name_1}' (Should be found) ---")
result_1 = binary_search_by_name(ranked_employees, search_name_1)
print(result_1)

print(f"\n--- Searching for '{search_name_2}' (Should be found) ---")
result_2 = binary_search_by_name(ranked_employees, search_name_2)
print(result_2)

print(f"\n--- Searching for '{search_name_3}' (Should be None) ---")
result_3 = binary_search_by_name(ranked_employees, search_name_3)
print(result_3)
```
