# Product Catalogue Cleaner
## Problem Statement

A supplier sends a raw product list with mixed types, duplicates, and missing values.  
Clean the data, convert all prices to float, group products into a dictionary by category,  
and print a neat summary.

---
- Step 1:

```python

raw_products = [
    {"name": "Laptop", "price": "120000", "category": "Electronics"},
    {"name": "Mobile", "price": 50000, "category": "Electronics"},
    {"name": "Shirt", "price": "2500", "category": "Clothing"},
    {"name": "Laptop", "price": "120000", "category": "Electronics"},
    {"name": "TV", "price": None, "category": "Electronics"},        
    {"name": "Shoes", "price": "abc", "category": "Clothing"},        
    {"name": "Watch", "price": 3000, "category": None},                
]
```

- Step 2: Data Cleaning Logic & Type Casting

```python
clean_products = []
seen_products = set()  

for product in raw_products:
    
    name = product.get("name")
    price = product.get("price")
    category = product.get("category")
    if name is None or price is None or category is None:
        continue
    try:
        price = float(price)
    except (ValueError, TypeError):
        continue  
    
    key = (name.lower(), price, category.lower())
    
    if key not in seen_products:
        seen_products.add(key)
        clean_products.append({
            "name": name.strip().title(),
            "price": price,
            "category": category.strip().title()
        })

clean_products
```

- Step 3: Group Products by Category

```python
grouped_products = {}

for product in clean_products:
    category = product["category"]
    grouped_products.setdefault(category, [])
    grouped_products[category].append(product)

grouped_products
```

- Step 4: Print neat Summary 

```python
print("CLEAN PRODUCT CATALOGUE SUMMARY\n")

for category, products in grouped_products.items():
    print(f"Category: {category}")
    print("-" * 30)
    
    for p in products:
        print(f"Name: {p['name']} | Price: {p['price']}")
    
    print()
```
